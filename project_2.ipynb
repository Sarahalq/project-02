```python
page = response.text
```

```python
# if needed: pip install beautifulsoup4 lxml or conda install beautifulsoup4 lxml
from bs4 import BeautifulSoup

soup = BeautifulSoup(page, "lxml")
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.find('a')
```

```python
soup.find('a',attrs={'target':"_blank"})
```

```python
for element in soup.find_all(class_='content-block'):
    print(element.prettify(), '\n')
```

```python
for element in soup.find_all(class_='title'):
    print(element.prettify(), '\n')
```
